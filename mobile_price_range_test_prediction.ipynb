# predict mobile price range from test dataset using model 

```python
import pandas as pd
import joblib
```

```python
test_df = pd.read_csv("datasets/test.csv")
```

```python
test_df.head()
```

```python
test_df.shape
```

```python
x=test_df.drop('id',axis=1)
```

```python
scaler= joblib.load('scaler.h5')
model=joblib.load('model.h5')
```

```python
scaled_x_test = scaler.transform(x)
scaled_x_test
```

```python
predict_price_range = model.predict(scaled_x_test)
```

```python
predict_price_range
```

#### add prediction on the test_df 

```python
test_df['predicted_price_range'] = predict_price_range
test_df.head()
```

#### save prediction on new csv file 

```python
predict_test = test_df[['id','predicted_price_range']]
predict_test.head()
```

```python
predict_test.to_csv('datasets\predict_test.csv',index=False)
```
